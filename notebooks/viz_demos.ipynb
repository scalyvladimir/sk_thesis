```python
domain = 'siemens15'

path = f'/home/v_chernyy/CC359/{domain}'
```

```python
from PIL import Image
import glob
import numpy as np

rand_scan = np.random.choice(os.listdir(f'{path}/train'))

img_list = []
files = glob.glob (f'{path}/train/{rand_scan}/*.png')

for f in sorted(files, key=lambda x: int(x.split('/')[-1].split('.')[0][3:]), reverse=True):
    image = np.asarray(Image.open(f))
    
    img_list.append(image)

print('X_data shape:', np.array(img_list).shape)
```

```python
img_scan = np.array(img_list)[10:]
```
